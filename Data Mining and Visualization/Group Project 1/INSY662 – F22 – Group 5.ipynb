```python
# Preproc step 1
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
#Decreasing dataframe size to save processing time 
urev2=urev.iloc[0:1000,:]
```

```python
urev2
```

```python
#remove stopwords
pat = r'\b(?:{})\b'.format('|'.join(stop))
urev2['review_without_stopwords'] = urev2['text'].str.replace(pat, '', regex = True)
urev2['review_without_stopwords'] = urev2['review_without_stopwords'].str.replace(r'\s+', ' ', regex = True)
#remove punc
urev2['no_punc'] = urev2['review_without_stopwords'].str.replace('[^\w\s]', '')
#lowercase
urev2['final_rev'] = urev2['no_punc'].str.lower()
```

```python
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
lemmatizer = WordNetLemmatizer()
```

```python
#Tokenize and lemmatize reviews
urev2['lem'] = ''
for i in range(len(urev2)):
    tokens_words = nltk.word_tokenize(urev2['final_rev'][i])
    urev2['lem'][i] = ' '.join([lemmatizer.lemmatize(w) for w in tokens_words])
```

```python
#get polarity
import textblob
from textblob import TextBlob

urev2['polarity'] = ''
for i in range(len(urev2)): 
    urev2['polarity'][i] = TextBlob(urev2['lem'][i]).sentiment.polarity
```
