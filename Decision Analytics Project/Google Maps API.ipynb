```python
import pandas as pd
import googlemaps
import requests 
from pprint import pprint
```

```python
#reading API key 
f = open('C:\\Users\\m.maraqa\\Desktop\\MMA courses\\Fall 2022\\Decision Analytics\\Project\\Google Maps API Key.txt', 'r')
api_key = f.readline()
f.close()
```

```python
maps = googlemaps.Client(api_key)
```

```python
url = 'https://maps.googleapis.com/maps/api/distancematrix/json?units=imperial&' 
#adis = UofT, Loblaws MLG, NoFrills Joe's, Walmart, Grocery Gateway, Metro
origin = '27 King\'s College Cir, Toronto, ON M5S | 60 Carlton St., Toronto, ON M5B 1L1 | Rocco\'s NOFRILLS Toronto, 200 Front St E, Toronto, ON M5A 4T9| 345 Bloor St E Unit 1A Unit 1A, Toronto, ON M4W 3J6 | 225 Rexdale Blvd, Etobicoke, ON M9W 1P7 | Dufferin Mall Supercentre, 900 Dufferin St, Toronto, ON M6H 4A9 | Loblaws Queen Street, 585 Queen St W, Toronto, ON M5V 2B7 | 444 Yonge St, Toronto, ON M5G 2B3'
dest = '27 King\'s College Cir, Toronto, ON M5S | 60 Carlton St., Toronto, ON M5B 1L1 | Rocco\'s NOFRILLS Toronto, 200 Front St E, Toronto, ON M5A 4T9| 345 Bloor St E Unit 1A Unit 1A, Toronto, ON M4W 3J6 | 225 Rexdale Blvd, Etobicoke, ON M9W 1P7 | Dufferin Mall Supercentre, 900 Dufferin St, Toronto, ON M6H 4A9 | Loblaws Queen Street, 585 Queen St W, Toronto, ON M5V 2B7 | 444 Yonge St, Toronto, ON M5G 2B3'
mode = 'transit&transit_mode=subway'
#get response
response = requests.get(url+'origins='+origin+'&destinations='+dest+'&mode='+mode+'&key='+api_key) #defualt mode of transport is driving. Other modes will have to be passed in mode argument
```

```python
pprint(response.json()['rows'])
```

```python
org_dest = pd.DataFrame()
places = ['UofT','Loblaws MLG', 'No Frills', 'No Frills Joe\'s','Walmart', 'Grocery Gateway', 'Loblaws', 'Metro']
p = len(places)

Os = []
Ds = []
meters = []
sec = []
for o in range(p):
    for d in range(p):
        Os.append(places[o]) 
        Ds.append(places[d])
        if response.json()['rows'][o]['elements'][d]['status'] == 'ZERO_RESULTS':
            meters.append(0)
            sec.append(0)
        else:
            meters.append(response.json()['rows'][o]['elements'][d]['distance']['value'])
            sec.append(response.json()['rows'][o]['elements'][d]['duration']['value'])

org_dest['Origin'] = Os
org_dest['Destination'] = Ds
org_dest['Travel Distance (m)'] = meters
org_dest['Travel Duration (s)'] = sec

#export to excel 
org_dest.to_csv(r'C:\\Users\\m.maraqa\\Desktop\\MMA courses\\Fall 2022\\Decision Analytics\\Project\\Origin Destination Matrix_'+mode+'.csv', index = False)

org_dest
```
